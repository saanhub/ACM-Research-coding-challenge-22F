```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        
import matplotlib.pyplot as plt 


#Scatterplot showing value for money ratings over the years

def sctrplot():
    data = pd.read_csv(os.path.join(dirname,filename))

    x = np.array(data.Year)
    y = np.array(data.ValueForMoneyRating)

    plt.scatter(x,y)
    plt.xlabel("Year")
    plt.ylabel("Value For Money Rating")
    plt.title("Value for money ratings over the years")
    plt.show()

sctrplot()


#Plotting a graph between Years and Number of cars sold per year

def bargraph():
    data = pd.read_csv(os.path.join(dirname,filename))

    #Counting number of Years
    datayr=data.Year
    unyr=np.unique(datayr)

    #Counting number of cars sold in each year
    d={}
    for i in unyr:
        ct = np.count_nonzero(data.Year == i)
        d[i]=ct

    #Plotting a graph between Years and Number of cars sold per year

    dval=d.values()

    plt.bar(unyr,dval)
    plt.xlabel("Years")
    plt.ylabel("Number of cars sold")
    plt.title("Graph for number of cars sold per year")
    plt.show()
    
bargraph()



#Pie chart for sales of car makes

import operator

def piemake():
    #Counting number of Makes
    datamake=data.Make
    undm=np.unique(datamake)

    #Counting number of cars sold by each Make
    d={}
    for i in undm:
        ct = np.count_nonzero(data.Make == i)
        d[i]=ct        
        
    sorted_d = dict( sorted(d.items(), key=operator.itemgetter(1),reverse=True))     

    lv=list(sorted_d.values())
    l1=lv[0:5]
    
    lk=list(sorted_d.keys())
    l2=lk[0:5]

    #Pie chart with top 5 Make
    plt.pie(l1, labels=l2)
    plt.title("Sales of top 5 car companies")
    plt.show()   
    
piemake()

"""#Line graphs for DealTypes over the years

def dtyrs():     
    #Counting number of Makes
    datadt=str(data.DealType)
    undt=np.unique(datadt.strip())

    #Counting number of cars sold by each Make
    d={}
    for i in undt:
        ct = np.count_nonzero(data.DealType == i)
        d[i]=ct
    print(d)
    
dtrys()""" 

#The above segment will work provided the data is cleaned.
    

    
    



```
